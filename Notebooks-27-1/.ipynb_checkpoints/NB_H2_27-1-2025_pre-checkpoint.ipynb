# Stochastic Dynamical Models

Process 1: Gaussian process (white noise) 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T=100; N=10000;  
t=np.linspace(0,T,N)
```

```python
G = np.random.normal(0.0,1.0,N)

plt.plot(t,G)
plt.xlabel(r'$t$',fontsize=16); plt.ylabel(r'$G_t$',fontsize=16,rotation=0)
plt.title('A Gaussian process path')
plt.show()
```

```python
import scipy
from scipy.stats import skew,kurtosis
print(np.mean(G))
print(np.var(G))
print(skew(G, axis=0, bias=True))
print(kurtosis(G, axis=0, fisher=False, bias=True))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T=100; N=20000; dt=float(T)/N; 
t=np.linspace(0,T,N+1)
```

```python
dW=np.sqrt(dt)*np.random.normal(0.0,1.0,N)
W=np.cumsum(dW);
W=np.insert(W,0,0)
```

```python
plt.plot(t,W)
plt.xlabel(r'$t$',fontsize=16); plt.ylabel(r'$W_t$',fontsize=16,rotation=0)
plt.title('A Wiener process path')
plt.show()
```

```python
import scipy
from scipy.stats import skew,kurtosis
print(np.mean(dW))
print(np.mean(dW*dW), dt)
```
